```python
import json, os
from selenium import webdriver
from selenium.webdriver.edge.options import Options
```

```python
# Cell 1: launch Edge with your Default profile and remote debugging
import subprocess, time, os

EDGE_PATH = r"C:\Program Files (x86)\Microsoft\Edge\Application\msedge.exe"
USER_DATA = os.path.expandvars(r"%LOCALAPPDATA%\Microsoft\Edge\User Data")
PROFILE   = "Default"

subprocess.Popen([
    EDGE_PATH,
    f"--remote-debugging-port=9222",
    f"--user-data-dir={USER_DATA}",
    f"--profile-directory={PROFILE}"
])
time.sleep(3)   # give Edge a moment to start up
```

```python
# Cell 2: attach Selenium to that running Edge and open Google Docs

opts = Options()
opts.add_experimental_option("debuggerAddress", "127.0.0.1:9222")
driver = webdriver.Edge(options=opts)

driver.get("https://docs.google.com")
print("Attached to existing Edge – title:", driver.title)
```
